$ SOC_t =\int_{0}^{t} X_s ds - t \bar{X}$ 
Using trapezoidal rule, simulate SOCt, Xt

```python
import numpy as np
import matplotlib.pyplot as plt
def demandSimulate(α = 0.5, K0 = 0, σ=2, n_step=1000, n_sim=10000, maturity = 100 , P0=0):
    
    dt = maturity/n_step
    priceMatrix = np.zeros((n_sim, n_step+1))
    SOCMatrix = np.zeros((n_sim,n_step+1))
    priceMatrix[:,0] = np.ones(n_sim) * P0
    SOCMatrix[:,0] = 0
    dW = np.random.normal(0,1,size = (n_sim,n_step) ) * np.sqrt(dt)
    
    for i in range(1,n_step+1):
        priceMatrix[:,i] = priceMatrix[:,i-1] + α * (K0 - priceMatrix[:,i-1]) * dt + σ * dW[:,i-1]
        SOCMatrix[:,i] = SOCMatrix[:,i-1]+ dt * (priceMatrix[:,i-1]+priceMatrix[:,i])/2 - dt * K0
        
    return priceMatrix,SOCMatrix
```

Compute $SOC_{max}, X_{max}$ from t $\in [0,1]$

```python
def path_maximum(rev, drift):
    X_t, SOC_t = demandSimulate(α = rev , σ = drift)
    X_max = np.mean(np.max(X_t,axis = 1))
    SOC_max = np.mean(np.max(SOC_t,axis = 1))
    return X_max,SOC_max
```

```python
path_maximum(0.5,2)
```

Plot variation of $\alpha$ and $\sigma$ against max MC. Negative mean reversion rate is impossible. O.W., no longer OU process due to limit of mean and var tends to $\infty$.


```python
reversion_rates = np.arange(0,2.1,0.1)
drifts  = np.arange(0,2.1,0.1)
max_Xt = np.zeros((len(drifts),len(reversion_rates)))
max_SOCt = np.zeros((len(drifts),len(reversion_rates)))
for i in range(len(reversion_rates)):
    for j in range(len(drifts)):
        max_Xt[j,i],max_SOCt[j,i] = path_maximum(rev = reversion_rates[i] , drift= drifts[j])
        
# create 2-D grids for plotting
```

```python
X, Y = np.meshgrid(reversion_rates, drifts)

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d') 
ax.plot_surface(X, Y, max_Xt, cmap='cool', alpha=0.8)
ax.set_title('max X_t, t = [0,1] ', fontsize=14)
ax.set_xlabel('reversion rate', fontsize=12)
ax.set_ylabel('drift', fontsize=12)
ax.set_zlabel('max X_t', fontsize=12)
plt.show()
```

```python
X, Y = np.meshgrid(reversion_rates, drifts)

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d') 
ax.plot_surface(X, Y, max_SOCt, cmap='cool', alpha=0.8)
ax.set_title('max SOC_t, t = [0,1] ', fontsize=14)
ax.set_xlabel('reversion rate', fontsize=12)
ax.set_ylabel('drift', fontsize=12)
ax.set_zlabel('max SOC_t', fontsize=12)
plt.show()
```

$SOC_{t+ \delta} - SOC{t} = \frac{X_{t+\delta} + X_t}{2} \delta - \delta \bar{X}$
Then, $\frac{SOC_{t+ \delta} - SOC{t}}{\delta} = \frac{X_{t+\delta} + X_t}{2} -\bar{X}$

Take a = $\sigma e^{-a(t+\delta})$
b = $\sigma e^{-at}$
Assume starting at mean, $X_{t+\delta} + X_t$ ~ $N(2m, (a+b)^2 \frac{(e^{2\alpha t} -1)}{2\alpha} +(a)^2 \frac{(e^{2\alpha t+\delta} -e^{2\alpha t})}{2\alpha} )$


```python
def variance(rev,drift):
    t = 50
    dt = 0.001
    a =drift *np.exp(-rev *(t+dt))
    b = drift*np.exp(-rev*t)
    c = (np.exp(2*rev*t)-1)/(2*rev)
    d = (np.exp(2*rev*t+dt)-np.exp(2*rev*t))/(2*rev)
    return ((a+b)**2 *c +a**2 *d)/4
```

```python

reversion_rates = np.arange(0.1,5.1,0.01)
drifts  = np.arange(0.1,5.1,0.01)
variances = np.zeros((len(drifts),len(reversion_rates)))
for i in range(len(reversion_rates)):
    for j in range(len(drifts)):
        variances[j,i] = variance(rev = reversion_rates[i] , drift= drifts[j])
        
```

```python
X, Y = np.meshgrid(reversion_rates, drifts)

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d') 
ax.plot_surface(X, Y, variances, cmap='cool', alpha=0.8)
ax.set_title('Var (SOC_t+dt - SOC_t)/dt, T= 10', fontsize=14)
ax.set_xlabel('reversion rate', fontsize=12)
ax.set_ylabel('drift', fontsize=12)
ax.set_zlabel('Var SOC_t', fontsize=12)
plt.show()

# T =10, dt = 0.001, 
```
